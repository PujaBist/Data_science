```python
import pandas as pd
```

```python
df=pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.columns

```

```python
df=df.rename(columns={'Annual Income (k$)' : 'annual_income',
                               'Spending Score (1-100)': 'spend_score'})
```

```python
df.head()
```

```python
x=df[['annual_income' , 'spend_score']]

x.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter( x.annual_income ,x.spend_score )
plt.xlabel('Annual_Income')
plt.ylabel('Spend_Score')
plt.show()
```

```python
# cluster vaneko akai thau ma occur vaako
```

```python
from sklearn.cluster import KMeans
```

```python
 kmeans=KMeans(n_clusters=4)
kmeans.fit(x)
```

```python
label=kmeans.labels_
label
```

```python
plt.scatter(x.annual_income , x.spend_score , c=label)
plt.show()
```
