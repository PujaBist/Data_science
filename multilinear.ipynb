```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('Salary_Data.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
x=data[['Age']]
y=data.Salary
```

```python
xtrain , xtest, ytrain , ytest=train_test_split(x,y, test_size=0.2, random_state=7)
```

```python
linear_model=LinearRegression()
linear_model.fit(xtrain, ytrain)
```

```python
linear_model.intercept_
```

```python
linear_model.coef_
```

```python
ypred=linear_model.predict(xtrain)
#for make line we are allowed to privide train or test data
```

```python
#for lin plot use plt.polt(training_data,  yprediction)
```

```python
plt.scatter(data.Age, data.Salary)
plt.plot(xtrain, ypred, color='Red')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
ypred_test=linear_model.predict(xtest)
```

```python
plt.scatter(xtest, ytest)
plt.plot(xtest, ypred_test, color='Red')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
