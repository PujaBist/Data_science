```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('Diabetes.csv')
data.head()
```

```python
x=data[['Glucose','BloodPressure','DiabetesPedigreeFunction','Age' ]] 
```

```python
x
```

```python
y=data.Outcome
```

```python
y
```

```python
xtrain , xtest, ytrain, ytest=train_test_split(x, y, test_size=0.2, random_state=90)
```

```python
dmodel=DecisionTreeClassifier(max_depth=7, criterion)
dmodel.fit(xtrain , ytrain)
```

```python
#model Evaluation
y_pred=dmodel.predict(xtest)
```

```python
acc=accuracy_score(ytest, y_pred)
print(f"The accuracy of decision tree {acc:.2%}")
```
