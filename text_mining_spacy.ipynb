```python
import spacy
nlp = spacy.load("en_core_web_sm") #is used to load the pre-trained English model that you downloaded earlier using python -m spacy download en_core_web_sm.
```

```python
# Read the file 
with open ('Text_mining.txt' , 'r', encoding='utf-8') as file:
    content=file.read()
```

```python
print(content)
```

```python
#Tokenization
doc = nlp(content) #tokens=content.split()
```

```python
print(doc)
```

```python
tokens = [token.text for token in doc]
print("Tokens:", tokens)
```

```python
#Count words
from collections import Counter
freq = Counter(tokens) #frequency=FreqDist(tokens)
```

```python
freq
```

```python
# same in nltk and spacy 
most_common = freq.most_common(16) #frequency.most_common(10) 
print("Most common:", most_common)
```

```python
#4.Remove Punctuation
tokens_no_punct = [token.text for token in doc if not token.is_punct]
print("Without punctuation:", tokens_no_punct)
```

```python
punctuations = [token.text for token in doc if token.is_punct] # to find punctuations in spacy 
# for nltk string.punctuations
```

```python
punctuations
```

```python
#Remove stopwords
tokens_no_stop = [token.text for token in doc if not token.is_stop and not token.is_punct]
print("Without stopwords:", tokens_no_stop)
```

```python
# for printing stop_words in spacy
#stop_words=set(stopwords.words('english')) in nltk
stop=[token.text for token in doc if  token.is_stop and not token.is_punct]
stop
```

```python
#Lemmatization
lemmas = [token.lemma_ for token in doc if not token.is_punct and not token.is_stop]
print("Lemmas:", lemmas)
```

```python
#Pos tagging
for token in doc:
    print(f"{token.text}: {token.pos_}")
```
