```python
from sklearn.datasets import make_blobs
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.svm import SVC
```

```python
x, y=make_blobs(200 , centers=2 , cluster_std=1.2, random_state=7) #must be pass in makeblobs no of data required   centers=> kati ota class vako data cheayo define garney 
```

```python
x, y=make_blobs(200,2 , centers=2 , cluster_std=1.7, random_state=7) # feature_name =2 by default
```

```python
x                                y
x1           x2
```

```python
x
```

```python
x_df=pd.DataFrame(x, columns=['A' ,'B'])
x_df.head()
```

```python
plt.scatter(x_df.A , x_df.B, c=y)
plt.xlabel('A')
plt.xlabel('B')
plt.show()
```

```python
print(x_df.columns)
```

```python
svm=SVC(kernel='linear') #  kernel function kun use garney vner yah decalre garanu parcha
# hyperplane mide point ma banch if we use kernel= 'linear'
svm.fit(x_df , y)
```

```python
svm.predict([[-1.1 , 5.0]])
```
