```python
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
data=sns.load_dataset('iris')
```

```python
data.head()
```

```python
x=data.drop('species' , axis=1)
y=data.species
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(x, y, test_size=0.2, random_state=9)
```

```python
dmodel=DecisionTreeClassifier()
dmodel.fit(xtrain,ytrain)
```

```python
ypred=dmodel.predict(xtest)
```

```python
ypred
```

```python
#Test Accuracy
acc=accuracy_score(ytest, ypred)
print(f"Accuracy of testing data: {acc:.2%}")
```

```python
y=dmodel.predict(xtrain)
# train accuracy
acc1=accuracy_score(ytrain, y)
print(f"Accuracy of training data: {acc1:.2%}")
```

```python
# grid search  library
from sklearn.model_selection import GridSearchCV
```

```python
dmodeln=DecisionTreeClassifier()
```

```python
#  estimators=> decisiontreeclassifier
# param_grid=>parameter haru ko dictionary
# scoring vaneko accuracy haru
# cv => cross validation
```

```python
parameter={
    'max_depth': [3,4,5,6,7]} # max_depth is a keyword
grid_search=GridSearchCV(estimator=dmodeln,
                         param_grid=parameter, 
                         scoring='accuracy',
                         cv=5)
grid_search.fit(xtrain, ytrain)

```

```python
grid_search.best_estimator_ # cv=5 xh so max_depth=3 :
acc    acc1    acc2    acc3   acc4    => sabko average nikalne then sabae depth ko avg aaccuracy herrxh then it returns 
max_depth =4
acc    acc1    acc2    acc3   acc4 
continue.......
```
