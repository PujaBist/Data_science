```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('wine.csv')
df.head()
```

```python
Dimension reduced garney ways haru=>
Feature Selection -> yi vaako maddey take only subset(corrleation
PCA -> reduced the dimensions -> yo le new column banunch  -> principla component vaninch -> yi sabae column ko information
-> jasko PCA highest hunch tesko contribution high hunch
```

```python
Curse of dimensionaly
```

```python
df=df.drop('class' , axis=1)
```

```python
df.head()
```

```python
#SCaling
scaler=StandardScaler()
dfs=scaler.fit_transform(df)
```

```python
dfs=pd.DataFrame(dfs, columns=df.columns)
```

```python
dfs.head()
```

```python
#Applying PCA
pca=PCA()
df_pca=pca.fit_transform(dfs)
```

```python
df_pca
```

```python
df_pca=pd.DataFrame(df_pca,)
```

```python
df_pca.head()
```

```python
first column le sab vanda badi information 1 thwn 2 , 3, 4, 5,6,7..............
```

```python
# for finding optimal no of component
=> Kaiser Rule
=> Scree plot ->it used concept of eigen value and  eigen vector
each column ko afhno afhno eigen value hunch
```

```python
Ax= lamdax lamda is eigen value, x is eigen vector
det(A-lamdaI)=0  yesari lamda ko value nikalne  
=> lamda ko value=no of column
=> jun column ko higest eigen value ch tyo column mtr select hunch
=> PCA le eigen value
```

```python
#Eigen Values
eigen=pca.explained_variance_
eigen
```

```python
#KAiser Rule states that those column only select whose eigen value is greater than 1
```

```python
#Explained Variance
variance=pca.explained_variance_ratio_
variance
```

```python
# description of above output
0.36 => this columns explains 36% of information of above data
```

```python
#Scree plot
plt.plot(range(1,14),eigen)
plt.xlabel('No of  components')
plt.ylabel('Eigen values')
plt.show()
```
