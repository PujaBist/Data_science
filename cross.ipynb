```python
data=pd.read_csv('Diabetes.csv')
data.head()
```

```python
x=data[['Glucose' ,'Age']]
y=data.Outcome
```

```python
xtrain , xtest, ytrain, ytest=train_test_split(x, y, test_size=0.2, random_state=9)
```

```python
dmodel=DecisionTreeClassifier()
dmodel.fit(xtrain, ytrain)
```

```python
ypred=dmodel.predict(xtest)
```

```python
acc=accuracy_score(ytest, ypred)
print(f"Accuracy  {acc:.2%}")
```

# cross validation

```python
# testing ,  training
```

```python
training         testing 
accuracy is high in training data
```

```python
ypred_train=dmodel.predict(xtrain)
```

```python
acc_train=accuracy_score(ytrain, ypred_train)
print(f"Accuracy of training of data {acc_train:.2%}")
```

```python
training data have high accuraccy and testting data have low accuracy this is called overfiiting
```

```python
taining data and test data both have low accuracy this is called underfitting
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(dmodel, xtrain, ytrain,cv=5)
print(score)
```

```python
score1=cross_val_score(DecisionTreeClassifier(), x, y,cv=5)
print(score1)
```
